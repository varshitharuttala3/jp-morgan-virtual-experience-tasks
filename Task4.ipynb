```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/Task 3 and 4_Loan_Data (1).csv")
df.head()
```

```python
df = df[['fico_score', 'default']]
df = df.dropna()
```

```python
df = df.sort_values(by='fico_score')
```

```python
num_buckets = 5
```

```python
df['fico_bucket'] = pd.qcut(
    df['fico_score'],
    q=num_buckets,
    labels=False
)
```

```python
bucket_summary = df.groupby('fico_bucket').agg(
    min_fico=('fico_score', 'min'),
    max_fico=('fico_score', 'max'),
    default_rate=('default', 'mean'),
    count=('default', 'count')
)

bucket_summary
```

The FICO scores are bucketed into five groups using quantiles so that
each bucket contains a similar number of borrowers. For each bucket,
the default rate is calculated. Lower FICO buckets show higher default
rates, while higher FICO buckets show lower default rates. This allows
FICO scores to be used as a categorical indicator of credit risk.


```python
def assign_fico_bucket(score, table):
    for idx, row in table.iterrows():
        if row['min_fico'] <= score <= row['max_fico']:
            return idx
    return None
```
